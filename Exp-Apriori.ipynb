```python
dataset.rename(columns={"MILK,BREAD,BISCUIT":"PRODUCTS"},inplace=True)
```

```python
listProducts=list(dataset["PRODUCTS"].apply(lambda x:x.split(',')))
listProducts
```

```python
encode=TransactionEncoder()
encode_data=encode.fit(listProducts).transform(listProducts)
dataset=pd.DataFrame(encode_data,columns=encode.columns_)
dataset
```
