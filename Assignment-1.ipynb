```python
from sklearn.model_selection import train_test_split
```

```python
features = titanic[["Sex", "Age", "Pclass"]]
target = titanic["Survived"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.5, random_state=0)
```
