```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
tips=pd.read_csv("tips.csv")
tips
```

```python
x=tips["tip"]
y=tips["day"]
#plt.scatter(y,x)
plt.scatter(tips['day'],tips['tip'])
plt.xlabel("day")
plt.ylabel("tip")
plt.title("tips")
plt.show()
```

```python
plt.scatter(tips['day'],tips['tip'],tips['total_bill'],tips['size'])
plt.xlabel("day")
plt.ylabel("tip")
plt.title("tips")
#plt.legend('male','female')
plt.colorbar()
plt.show()
```

```python
plt.plot(tips['tip'])
plt.plot(tips['size'])
plt.xlabel("tip")
plt.ylabel("size")
plt.title("tips")
plt.show()
```

```python
plt.bar(tips['day'],tips['tip'])
plt.xlabel("day")
plt.ylabel("tip")
plt.title("tips")
plt.show()
```

```python
plt.hist(tips['total_bill'])
plt.xlabel("day")
plt.ylabel("tip")
plt.title("tips")
plt.show()
```
