```python
# Beautiful day at the movies problem statement.

def beautifulDays(i, j, k):
  beautifulDays = 0
  for day in range(i, j+1):
    reversedDay = int(str(day)[::-1])
    if abs(day - reversedDay) % k == 0:
      beautifulDays += 1
  return beautifulDays

print(beautifulDays(20, 23, 6))
```

```python
class Circle:
    def __init__(self, r):
        self.r = r

    def area(self):
        print(3.14 * self.r * self.r)

    def perimeter(self):
        print(2 * 3.14 * self.r)

    def diameter(self):
        print(2 * self.r)

# Creating an object of the Circle class with radius 4
obj = Circle(4)
obj.area()  # Calling area method
obj.perimeter()  # Calling perimeter method
obj.diameter()  # Calling diameter method
```

```python
class circle:
  def add(self,a,b):
    print(a+b)
  def sub(self,a,b):
    print(a-b)
  def mul(self,a,b):
    print(a*b)
  def div(self,a,b):
    print(a/b)
  def mod(self,a,b):
    print(a%b)
obj=circle()
obj.add(4,6)
obj.sub(4,6)
obj.mul(4,6)
obj.div(4,6)
obj.mod(4,6)
```

```python
class Area_Perimeter:
  def __init__(self, length, width):
    self.length = length
    self.width = width

  def area(self):
    return self.length * self.width

  def perimeter(self):
    return 2 * (self.length + self.width)

obj = Area_Perimeter(4, 6)

print("Area:", obj.area())
print("Perimeter:", obj.perimeter())
```

```python
class Student:
    def __init__(self, id):
        self.id = id

    def set_name(self, name):
        self.name = name

    def check(self, age):
        if age >= 18:
            print(f"{self.name} is eligible.")
        else:
            print(f"{self.name} is not eligible.")

ch = Student(1)
ch.set_name("John")
ch.check(21)
ch.check(10)
```
