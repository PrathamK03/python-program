```python
class cmp:
  def __init__(self,dept,address):
    self.dept=dept
    self.address=address
  def show(self):
    print("my dept is "+self.dept+" and my address is "+self.address)
company=cmp("finance","pune")
company.show()
```

```python
class company:
  def __init__(self,dept,address,mobile):
    self.dept=dept
    self.address=address
    self.mobile=mobile
  def __str__(self):
    return f"my dept is {self.dept} and my address is {self.address} and my mobile number is {self.mobile}"
company1=company("marketing","wakad","1234")
print(company1)
```
