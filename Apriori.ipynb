```python
import numpy as  np
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
retail=pd.read_excel('Online_Retail.xlsx')
retail.head()
```

```python
retail.columns
```

```python
retail.Country.unique()
```

```python
retail['Description']=retail['Description'].str.strip()
#Removing extra spaces in desc field
```

```python
retail.dropna(axis=0,subset=['InvoiceNo'],inplace=True)
retail['InvoiceNo']=retail['InvoiceNo'].astype('str')
```

```python
retail=retail[~retail['InvoiceNo'].str.contains('C')]
basket_france=(retail[retail['Country']=="France"].groupby(['InvoiceNo','Description'])['Quantity'].sum().unstack().reset_index().fillna(0).set_index('InvoiceNo'))
```

```python
basket_uk=(retail[retail['Country']=="United Kingdom"].groupby(['InvoiceNo','Description'])['Quantity'].sum().unstack().reset_index().fillna(0).set_index('InvoiceNo'))
```

```python
basket_sweden=(retail[retail['Country']=="Sweden"].groupby(['InvoiceNo','Description'])['Quantity'].sum().unstack().reset_index().fillna(0).set_index('InvoiceNo'))
```

```python
basket_eire=(retail[retail['Country']=="EIRE"].groupby(['InvoiceNo','Description'])['Quantity'].sum().unstack().reset_index().fillna(0).set_index('InvoiceNo'))
```

```python
basket_belgium=(retail[retail['Country']=="Belgium"].groupby(['InvoiceNo','Description'])['Quantity'].sum().unstack().reset_index().fillna(0).set_index('InvoiceNo'))
```

```python
def hot_encode(x):
  if x<=0:
    return 0
  if x>=1:
    return 1
basket_encoded=basket_france.applymap(hot_encode)
basket_france=basket_encoded
basket_encoded=basket_uk.applymap(hot_encode)
basket_uk=basket_encoded
basket_encoded=basket_sweden.applymap(hot_encode)
basket_sweden=basket_encoded
basket_encoded=basket_belgium.applymap(hot_encode)
basket_belgium=basket_encoded
```

```python
freq_items=apriori(basket_france,min_support=0.05,use_colnames=True)
rules=association_rules(freq_items,metric="lift",min_threshold=1)
rules=rules.sort_values(['confidence','lift'],ascending=[False,False])
print(rules.head())
```
