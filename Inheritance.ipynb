```python
# Single/Simple inheritance.

# Parent class.

class person:
  def __init__(self,name,idnum):
    self.name=name
    self.idnum=idnum
  def display(self):
    print(self.name)
    print(self.idnum)
  def details(self):
    print("my name is {}".format(self.name))
    print("my id number is {}".format(self.idnum))
  def pqr(self):
    pass

# Child class.

class employee(person):
  def __init__(self,name,idnum,salary,post):
    self.salary=salary
    self.post=post
    person.__init__(self,name,idnum)
  def details(self):
    print("my name is {}".format(self.name))
    print("my id number is {}".format(self.idnum))
    print("my salary is {}".format(self.salary))
    print("my post is {}".format(self.post))

abc=employee('xyz',1,20000,'intern')
abc.display()
abc.details()
abc.pqr()
```

```python
# Multiple inheritance.

# Parent class.

class person:
  def __init__(self,name,idnum):
    self.name=name
    self.idnum=idnum
  def display(self):
    print(self.name)
    print(self.idnum)
  def details(self):
    print("my name is {}".format(self.name))
    print("my id number is {}".format(self.idnum))

# Child class.

class employee(person):
  def __init__(self,name,idnum,salary,post):
    self.salary=salary
    self.post=post
    person.__init__(self,name,idnum)
  def details(self):
    print("my name is {}".format(self.name))
    print("my id number is {}".format(self.idnum))
    print("my salary is {}".format(self.salary))
    print("my post is {}".format(self.post))

class manager(employee):
  def __init__(self,name,idnum,salary,post,mobno,email):
    self.mobno=mobno
    self.email=email
    employee.__init__(self,name,idnum,salary,post)
  def show(self):

    print("my mobile no is {}".format(self.mobno))
    print("my email is {}".format(self.email))

abc=manager('xyz',1,20000,'intern',1234,'xyz@gmail.com')
#abc.display()
#abc.details()
abc.show()
```

```python
# Hybrid inheritance.

# Parent class.

class person:
  def __init__(self,name,idnum):
    self.name=name
    self.idnum=idnum
  def display(self):
    print(self.name)
    print(self.idnum)
  def details(self):
    print("my name is {}".format(self.name))
    print("my id number is {}".format(self.idnum))

# Child 1 class.

class employee1(person):
  def __init__(self,name,idnum,salary,post):
    self.salary=salary
    self.post=post
    person.__init__(self,name,idnum)
  def details(self):
    print("my name is {}".format(self.name))
    print("my id number is {}".format(self.idnum))
    print("my salary is {}".format(self.salary))
    print("my post is {}".format(self.post))

# Child 2 class.

class employee2(person):
  def __init__(self,name,idnum,address,nationality):
    self.address=address
    self.nationality=nationality
    person.__init__(self,name,idnum)
  def details(self):
    print("my name is {}".format(self.name))
    print("my id number is {}".format(self.idnum))
    print("my address is {}".format(self.address))
    print("my nationality is {}".format(self.nationality))

# Next class.

class manager(employee1,employee2):
  def __init__(self,name,idnum,mobno,email):
    self.mobno=mobno
    self.email=email
    person.__init__(self,name,idnum)
  def show(self):

    print("my name is {}".format(self.name))
    print("my id number is {}".format(self.idnum))
    print("my mobile no is {}".format(self.mobno))
    print("my email is {}".format(self.email))

abc=manager('xyz',1,20000,'intern')

abc.display()
#abc.details()
abc.show()

abc1=employee('abc',2,30000,'manager')
abc1.display()
abc1.details()
```
