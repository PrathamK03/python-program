```python
# Performing matrix operations using array.

import numpy as np

x1=[]
x=int(input('enter the number of elements:'))
for i in range(0,x):
  x=int(input())
  x1.append(x)
print(x1)

x2=[]
y=int(input('enter the number of elements:'))
for i in range(0,y):
  y=int(input())
  x2.append(y)
print(x2)

x3=[]
print('1:Addition')
print('2:Multiplication')
print('3:Substraction')
print('4:Division')
z=int(input('Enter the operation you want to perform:'))

if(z == 1):
  x3=np.add(x1,x2)
  print(x3)

elif(z == 2):
  x3=np.multiply(x1,x2)
  print(x3)

elif(z ==3):
  x3=np.subtract(x1,x2)
  print(x3)

elif(z == 4):
  x3=np.divide(x1,x2)
  print(x3)

else:
  print('invalid choice')
```
