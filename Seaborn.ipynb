```python
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
def sinplot(flip=1):
  x=np.linspace(0,10,100)
  for i in range(1,5):
    plt.plot(x,np.sin(x+i*0.5)*(10-i)*flip)
sb.set_style("darkgrid",{'axes.axisbelow':False})
sb.set_palette("Reds")
sinplot()
plt.show()
print(sb.axes_style())
```
