```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('esd.csv')
#plt.title('scatter plot')
df1=df[df['label']=='yes']
df2=df[df['label']=='no']
df3=df[df['label']=='neutral']
a=df1.plot(x='x axis',y='y axis',kind='scatter',c='r',label='yes')
df2.plot(x='x axis',y='y axis',kind='scatter',ax=a,c='b',label='no')
df3.plot(x='x axis',y='y axis',kind='scatter',ax=a,c='g',label='neutral')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
df=pd.DataFrame({'name':['a','a','a','a','b','b','b','b','c','c','c','d','d','d'],'votes':[10,5,12,5,8,8,1,4,5,4,3,7,8,9]})
df.groupby(['name']).sum().plot(kind='pie',y='votes',autopct='%1.0f%%')
```
