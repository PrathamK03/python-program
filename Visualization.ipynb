```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4]
y=[8,5,4,7]
plt.plot(x,y)
plt.show()
```

```python
x=[1,2,3,4]
y=[8,9,4,6]
plt.scatter(x,y,marker="*")
plt.show()
```

```python
x=[1]
y=[2]
plt.xlim(0,5)
plt.ylim(0,5)
plt.plot(x,y,marker="o",markersize=10,markeredgecolor="black")
plt.show()
```

```python
x=[4,3]
y=[1,2]
plt.plot(y,x)
plt.show()
```

```python
x=[1,2,3,4]
y=[3,5,3,1]
plt.boxplot(x,y)
plt.show()
```

```python
x=[1,2,3,4]
y=[3,5,3,1]
plt.bar(x,y)
plt.show()
```

```python
x=[1,2,3,4]
lab=["a","b","c","d"]
plt.pie(x,labels=lab)
plt.legend(title="pie chart")
plt.show()
```
