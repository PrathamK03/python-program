```python
import numpy as np
x=np.array([[1,2,3,4],[5,6,7,8]])
print(x)
print(np.zeros((2,3)))
print(np.ones((1,3)))
print(np.arange(7))
print(np.arange(0,7,0.4))
print(np.linspace(0.,7.,6))
print(np.matrix('1 2; 3 4'))
```
