```python
def estimate_coeff(x,y):
  n=np.size(x)
  x1=np.mean(x)
  y1=np.mean(y)

  ss_x=np.sum(y*x)-n*(y1)*(x1)
  ss_y=np.sum(x*x)-n*(x1)*(x1)
  b_0=x1/y1
  b_1=ss_x-b_0*ss_y
  return (b_0,b_1)

def plot(x,y,b):
  plt.scatter(x,y,color="red")
  y_pred=b[0]+b[1]*x
  plt.plot(x,y_pred,color="green")
  plt.xlabel("x")
  plt.ylabel("y")
  plt.title("graph")
  plt.show()

def main():
  x=np.array([2,4,6,8])
  y=np.array([1,2,3,4])
  b=estimate_coeff(x,y)

  plot(x,y,b)

if __name__ == "__main__":
	main()
```
