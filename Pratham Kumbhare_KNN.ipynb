```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
iris = load_iris()
iris
```

```python
import pandas as pd
df = pd.DataFrame(data= iris.data, columns = ["sepal_length", "sepal_width", "petal_length", "petal_width"])
df['species'] = iris.target
df.head()
```

```python
df.describe(include='all')
```

```python
# now split the data into x train .ytest
X = df.iloc[:,:4]
X.head()
```

```python
y = df.iloc[:,-1]
y.head()
```

```python
# now normalize the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
# train test split the data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3,random_state =1)
# 0.3 - out of 100 record 70 is for training  and 30 is for testing
# randaom_state = 1 means it will take make data again and again same selection will be used
y_test.shape # 45 rows are testing
```

```python
# now make the model
knnmodel = KNeighborsClassifier(n_neighbors=3) # 3- value of K
# now fit the model with the data
knnmodel.fit(X_train,y_train)
# after traing is done you'll use the model to predict the value
y_predict1=knnmodel.predict(X_test)
# now find the Accuracy of model
from sklearn.metrics import accuracy_score
acc = accuracy_score(y_test,y_predict1)
acc
```

```python
# this means tht 3% of data is missclassified
#now make a confusion matrix to represent the data
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test.values, y_predict1)  #cm is found on actual class label and predicted one
cm
```

```python
cm1 = pd.DataFrame(data = cm,index=['setosa','versicolor','virginia'],columns=['setosa','versicolor','virginia'])
cm1
```
