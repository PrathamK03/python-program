```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,3,6,2,1,8,7]
y=[2,4,6,8,1,3,5,6,3,1]
plt.scatter(x,y)
plt.show()
```

```python
from sklearn.cluster import KMeans
data=list(zip(x,y))
z=[]
for i in range(1,10):
  kmeans=KMeans(n_clusters=i)
  kmeans.fit(data)
  z.append(kmeans.inertia_)
plt.plot(range(1,10),z,marker='o')
plt.title('elbow method')
plt.xlabel('no of clusters')
plt.ylabel('z')
plt.show()
```

```python
kmeans=KMeans(n_clusters=4)
kmeans.fit(data)
plt.scatter(x,y,c=kmeans.labels_)
plt.show()
```
