```python
import numpy as np
import pandas as pd

# Uncomment this line if you want to read data from a CSV file
# iris_data = pd.read_csv('iris.csv')

# Assuming 'sepal_length' and 'sepal_width' are columns in the loaded data
# Replace this with the actual column names from your dataset
sepal_length = iris_data['sepal_length']
sepal_width = iris_data['sepal_width']

def correlation(x, y):
    mean_x = sum(x) / float(len(x))
    mean_y = sum(y) / float(len(y))
    sub_x = [i - mean_x for i in x]
    sub_y = [i - mean_y for i in y]
    numerator = sum([sub_x[i] * sub_y[i] for i in range(len(sub_x))])
    std_dev_x=sum([sub_x[i]**2.0 for i in range(len(sub_x))])
    std_dev_y=sum([sub_y[i]**2.0 for i in range (len(sub_y))])
    denominator = (std_dev_x*std_dev_y)**0.5
    corr = numerator / denominator
    return corr

# Example usage
corr_func = correlation(sepal_length, sepal_width)
print("correlation from the function:", corr_func)
```
