```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
creditcard_data=pd.read_csv('creditcard.csv')
creditcard_data
```

```python
creditcard_data.head()
```

```python
data_info=creditcard_data.info()
```

```python
creditcard_data.describe()
```

```python
fig,ax=plt.subplots(1,2,figsize=(18,6))
sns.histplot(creditcard_data['Time'],bins=50,ax=ax[0],kde=True)
ax[0].set_title('Distribution of Transaction Time')
ax[0].set_xlabel('Time (seconds)')

sns.histplot(creditcard_data['Amount'],bins=50,ax=ax[1],kde=True)
ax[1].set_title('Distribution of Transaction Amount')
ax[1].set_xlabel('Amount')

ax[0].set_ylabel('Frequency')
ax[1].set_ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Class',data=creditcard_data)
plt.title('Class Distribution (0: Non-Fraud, 1: Fraud)')
plt.xlabel('Class')
plt.ylabel('Frequency')
plt.show()
creditcard_data['Class'].value_counts()
```

```python
creditcard_data['Class'].value_counts().plot(kind='pie',autopct='%1.2f%%',explode=(0,0.1))
plt.title('Class Distribution')
plt.show()
```

```python
fig,ax=plt.subplots(1,2,figsize=(18,6))

#ax=ax[0]

sns.boxplot(x='Class',y='Amount',data=creditcard_data,ax=ax[0])
ax[0].set_title('Box Plot of Transaction Amount by Class')
ax[0].set_yscale('log')
ax[0].set_ylabel('Amount (log scale)')
ax[0].set_xlabel('Class')

sns.boxplot(x='Class',y='Time',data=creditcard_data,ax=ax[1])
ax[1].set_title('Box Plot of Transaction Time by Class')
ax[1].set_ylabel('Time (seconds)')
ax[1].set_xlabel('Class')
plt.tight_layout()
plt.show()
```

```python
corr_matrix=creditcard_data.corr()
plt.figure(figsize=(15,10))
sns.heatmap(corr_matrix,cmap='gray',annot=True,fmt=".1f")
plt.title('Correlation Heatmap of Features')
plt.show()
```
