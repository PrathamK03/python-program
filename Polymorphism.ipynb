```python
class bird:
  def intro(self):
    print('there are many types of birds')
  def flight(self):
    print('most birds can fly but some cannnot')

class sparrow(bird):
  def flight(self):
    print('sparrows can fly')

class ostrich(bird):
  def flight(self):
    print('ostriches cannot fly')

x=bird()
y=sparrow()
z=ostrich()

x.intro()
x.flight()

y.intro()
y.flight()

z.intro()
z.flight()
```

```python
class bird:
  def intro(self):
    print('there are many types of birds')

class bird1:
  def intro(self,name):
    self.name=name
  def show(self):
    print(self.name)

class bird2:
  def intro(self,addr,mob):
    self.addr=addr
    self.mob=mob
  def show(self):
    print(self.addr)
    print(self.mob)

x=bird()
y=bird1()
z=bird2()

x.intro()
y.intro('abc')
y.show()
z.intro('pune',123)
z.show()


#x.intro()
#x.intro('abc')
#x.intro('pune',123)
#x.show()
```
