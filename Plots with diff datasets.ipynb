```python
import matplotlib.pyplot as plt
x1=[1,2,3,4,5]
y1=[6,7,8,9,10]

x2=[2,4,6,3,1]
y2=[8,3,1,5,6]
fig,xy=plt.subplots()
xy.scatter(x1,y1,label='dataset1')
xy.scatter(x2,y2,label='dataset2',color='red')
xy.legend()
xy.set_xlabel('xaxis')
xy.set_ylabel('yaxis')
plt.show()
```

```python
import matplotlib.pyplot as plt
x1=[1,2,3,4,5]
y1=[6,7,8,9,10]

x2=[2,4,6,3,1]
y2=[8,3,1,5,6]

x3=[9,7,5,8,1]
y3=[3,1,6,3,4]
fig,xy=plt.subplots()
xy.scatter(x1,y1,label='dataset1')
xy.scatter(x2,y2,label='dataset2',color='red')
xy.scatter(x3,y3,label='dataset3',color='green')
xy.legend(loc='upper right')
xy.set_xlabel('xaxis')
xy.set_ylabel('yaxis')
plt.show()
```
