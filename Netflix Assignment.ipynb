```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
netflix=pd.read_csv('netflix_titles.csv')
netflix
```

 Make a visualization showing the total number of movies watched by children

```python
children=netflix['listed_in'].str.contains("Children & Family Movies",case=False)
kids=netflix['listed_in'].str.contains("Kids' TV",case=False)
print(f"filter:{type(kids)}\n{kids}",f"filter:{type(children)}\n{children}")
#print(f"filter:{type(children)}\n{children}")
```

```python
df_kids=netflix[children|kids]
df_kids.head()
```

Make a visualization showing the total number of standup comedies

```python
netflix.listed_in.unique()
```

```python
standup=netflix['listed_in'].str.contains("Stand-Up Comedy",case=False)
print(f"standup:{type(standup)}\n{standup}")
```

```python
df_standup=netflix[standup]
df_standup.head()
```

Make a visualization showing most watched shows

```python
import plotly.express as px
fig_donut=px.pie(netflix,names='type',height=300,width=600,title='most watched shows')
fig_donut.update_traces(hovertemplate=None,textposition='inside',textinfo='percent+label')
```

Make a visualization showing highest rated show

```python
plt.figure(figsize=(12,10))
ax=sns.countplot(x="rating",data=netflix,palette="Set2",order=netflix['rating'].value_counts().index)
```
